```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt 
```

```python
img = tf.keras.preprocessing.image.load_img("Abyssinian_1.jpg",target_size=(200,300))
plt.imshow(img)
```

```python
datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
)
```

```python
type(img)
```

```python
img = tf.keras.preprocessing.image.img_to_array(img)
type(img)
```

```python
img.shape
```

```python
img = img/255
```

```python
# we need to tell how many image we are giving that's why we need to reshpae()
input_batch = img.reshape(1,200,300,3)
```

```python
datagen.fit(input_batch)
batches = datagen.flow(input_batch,batch_size=10,save_to_dir="dataAugmen")
img = next(batches)
```
