```python
from gensim.models import Word2Vec,keyedvectors
```

```python
corpus = [
    ["machine", "learning", "is", "amazing"],
    ["learning", "about", "natural", "language", "processing", "is", "fun"],
    ["machine", "learning", "and", "deep", "learning", "are", "popular", "topics"],
    ["good","better","best"]
]

model = Word2Vec(sentences=corpus,vector_size=100,window=5,min_count=1,workers=4)

print("<------------- vector for learning: ------------------------->")
print(model.wv['learning'])
print(model.wv['learning'].shape)

print()
print("<------------- Most similar words: ------------------------->")
print(model.wv.most_similar(["good"])) # machine learning is good. we will get machine.
# The issue arises because Word2Vec determines word similarity based on the context 
# in which words appear, not their semantic meaning. That's why we don't get better an best.

model.save("word2vec.model")
```

```python


model = keyedvectors.load_word2vec_format("GoogleNews-vectors-negative300.bin.gz",binary=True,limit=50000)
```

```python
embedding_array = model["king"] - model["man"] + model["women"]
embedding_array
```

```python
for words,similarity_score in model.similar_by_vector(embedding_array,topn=5):
    print(f"similar word: {words}, Similarity: {similarity_score}")
```
